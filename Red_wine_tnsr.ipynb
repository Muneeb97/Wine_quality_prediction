```python
#Classification Problem
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('winequality-red.csv',sep=';')
df.rename(columns = {'fixed acidity':'fixedacidity'}, inplace = True) 
df.rename(columns = {'volatile acidity':'volatileacidity'}, inplace = True) 
df.rename(columns = {'citric acid':'citricacid'}, inplace = True) 
df.rename(columns = {'residual sugar':'residualsugar'}, inplace = True) 
df.rename(columns = {'free sulfur dioxide':'freesulfurdioxide'}, inplace = True) 
df.rename(columns = {'total sulfur dioxide':'totalsulfurdioxide'}, inplace = True)
```

```python
df.head()
```

```python
cols_normalize = ['fixedacidity', 'volatileacidity', 'citricacid', 'residualsugar',
       'chlorides', 'freesulfurdioxide', 'totalsulfurdioxide', 'density',
       'pH', 'sulphates', 'alcohol']
```

```python
#We can use this function to normalize our data from a real-life datasets
df[cols_normalize] = df[cols_normalize].apply(lambda x : (x - x.min())/(x.max() - x.min()))
```

```python
df.head()
```

```python
df.groupby('quality').count()
```

```python
#Inizializing Features
#Numerical Values
fixed_acidity = tf.feature_column.numeric_column('fixedacidity')
volatile_acidity =  tf.feature_column.numeric_column('volatileacidity')
citric_acid = tf.feature_column.numeric_column('citricacid')
residual_sugar = tf.feature_column.numeric_column('residualsugar')
chlorides = tf.feature_column.numeric_column('chlorides')
free_sulfur_dioxide = tf.feature_column.numeric_column('freesulfurdioxide')
total_sulfur_dioxide = tf.feature_column.numeric_column('totalsulfurdioxide')
density = tf.feature_column.numeric_column('density')
pH = tf.feature_column.numeric_column('pH')
sulphates = tf.feature_column.numeric_column('sulphates')
alcohol = tf.feature_column.numeric_column('alcohol')
```

```python
#,citric_acid
feat_columns = [fixed_acidity,volatile_acidity,residual_sugar,chlorides,free_sulfur_dioxide,total_sulfur_dioxide,
               density,pH,sulphates,alcohol]
```

```python
#Split our data for train test
X = df.drop('quality',axis = 1)
y = df['quality']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
#lets create our functions for training and testing
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train,
                                                 y=y_train,
                                                batch_size=10,
                                                num_epochs=1000,
                                                shuffle=True)
```

```python
dense_nn_model = tf.estimator.DNNClassifier(hidden_units=[10,10,10,10,10,6],
                                            feature_columns=feat_columns,
                                           n_classes=6,label_vocabulary=['3','4','5','6','7','8'])
```

```python
dense_nn_model.train(input_fn=input_func,steps = 1000)
```

```python
eval_input_fnc = tf.estimator.inputs.pandas_input_fn(x=X_test,
                                                    y=y_test,
                                                    batch_size=10,
                                                    num_epochs=1,
                                                    shuffle=False)
```

```python
pred_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test,
                                                     batch_size=10,
                                                     num_epochs=1,
                                                     shuffle=False)
```

```python
DNN_eval = dense_nn_model.evaluate(eval_input_fnc)
```

```python
DNN_eval
```

```python
Predictions_dnn = dense_nn_model.predict(pred_input_func)
```

```python
conv_predictions_dnn = list(Predictions_dnn)
```

```python
predict_class_id_dnn = []
for i in conv_predictions_dnn:
    for j in i['class_ids']:
        predict_class_id_dnn.append(j)
```

```python
original_test_data = []
for i in y_test:
    original_test_data.append(i)
```

```python
print('original,predict')
for i,j in zip(original_test_data,predict_class_id_dnn):
    print(i,'  ',j)
```
